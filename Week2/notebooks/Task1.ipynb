## Handling Missing Data

```python
data.head()
```

 Fill numeric columns with the mean and Fill non-numeric columns with the mode

```python
def fill_missing_values(Data):
    for column in data.columns:
        if pd.api.types.is_numeric_dtype(data[column]):
            # Fill numeric columns with the mean
            mean_value = data[column].mean()
            data[column].fillna(mean_value, inplace=True)
        else:
            # Fill non-numeric columns with the mode
            mode_value = data[column].mode()[0]  # Get the first mode
            data[column].fillna(mode_value, inplace=True)

# Apply the function
fill_missing_values(data)
```

```python
aggregated_data = data.groupby('IMSI').agg(
    number_of_sessions=('IMSI', 'count'),
    total_session_duration=('Dur. (ms)', 'sum'),
    total_download_data_bytes=('HTTP DL (Bytes)', 'sum'),
    total_social_media_download_bytes=('Social Media DL (Bytes)', 'sum'),
    total_upload_data_bytes=('HTTP UL (Bytes)', 'sum'),
    total_social_media_upload_bytes=('Social Media UL (Bytes)', 'sum'),
).reset_index()
```

```python

# Calculate total download and upload data
aggregated_data['total_download_data_bytes'] = (
    aggregated_data['total_download_data_bytes'] + 
    aggregated_data['total_social_media_download_bytes']
)
aggregated_data['total_upload_data_bytes'] = (
    aggregated_data['total_upload_data_bytes'] + 
    aggregated_data['total_social_media_upload_bytes']
)
```

```python
data.columns = data.columns.str.strip()

# Aggregation per user (MSISDN/Number)
user_aggregation = data.groupby('MSISDN/Number').agg({
    'Bearer Id': 'count',  # Number of xDR sessions
    'Dur. (ms)': 'sum',  # Total session duration (in milliseconds)
    'Total DL (Bytes)': 'sum',  # Total download data
    'Total UL (Bytes)': 'sum',  # Total upload data
    'Social Media DL (Bytes)': 'sum',
    'Social Media UL (Bytes)': 'sum',
    'Google DL (Bytes)': 'sum',
    'Google UL (Bytes)': 'sum',
    'Email DL (Bytes)': 'sum',
    'Email UL (Bytes)': 'sum',
    'Youtube DL (Bytes)': 'sum',
    'Youtube UL (Bytes)': 'sum',
    'Netflix DL (Bytes)': 'sum',
    'Netflix UL (Bytes)': 'sum',
    'Gaming DL (Bytes)': 'sum',
    'Gaming UL (Bytes)': 'sum',
    'Other DL (Bytes)': 'sum',
    'Other UL (Bytes)': 'sum'
}).reset_index()

user_aggregation.head()
```

```python
# chckin Outliers 
from scipy import stats
import numpy as np
# Calculate Z-scores
numeric_cols = user_aggregation.select_dtypes(include=[np.number]).columns
z_scores = stats.zscore(user_aggregation[numeric_cols])
abs_z_scores = np.abs(z_scores)
user_aggregation_cleaned = user_aggregation[(abs_z_scores < 3).all(axis=1)] 
```

### Variable Transformation

```python
# Segment users into top five decile classes based on total session duration
user_aggregation_cleaned['Decile'] = pd.qcut(user_aggregation_cleaned['Dur. (ms)'], 10, labels=False)
decile_summary = user_aggregation_cleaned.groupby('Decile').agg({
    'Total DL (Bytes)': 'sum',
    'Total UL (Bytes)': 'sum'
})
decile_summary.head()
```

```python
# Basic Metrix

metrics_summary = user_aggregation_cleaned.describe()
metrics_summary
```

```python
# Univarient Analysis 
# Compute dispersion parameters
numeric_cols = user_aggregation_cleaned.select_dtypes(include=[np.number]).columns
for column in numeric_cols:
    print(f'{column} - Variance: {user_aggregation_cleaned[column].var()}, Standard Deviation: {user_aggregation_cleaned[column].std()}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot histograms for each numerical column
for column in numeric_cols:
    plt.figure()
    sns.histplot(user_aggregation_cleaned[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

### Bivraiant Analysis

```python
# Analyze relationship between each application and Total DL + UL data
user_aggregation_cleaned['Total Data'] = user_aggregation_cleaned['Total DL (Bytes)'] + user_aggregation_cleaned['Total UL (Bytes)']

applications = [
    'Social Media DL (Bytes)', 'Google DL (Bytes)', 'Email DL (Bytes)',
    'Youtube DL (Bytes)', 'Netflix DL (Bytes)', 'Gaming DL (Bytes)', 'Other DL (Bytes)'
]

for app in applications:
    plt.figure()
    sns.scatterplot(x=user_aggregation_cleaned[app], y=user_aggregation_cleaned['Total Data'])
    plt.title(f'{app} vs Total Data')
    plt.xlabel(app)
    plt.ylabel('Total Data')
    plt.show()
```
